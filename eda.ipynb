Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import sklearn as sk
import numpy as np
import pandas as pd
import spici
from spici import dtex
%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
# plt.figure(figsize = (0.5, 0.5))
```

```python
dataset = pd.read_csv('train.csv')
data = dataset.copy()
data.drop('Id', axis = 1, inplace = True)
```

```python
print(dataset.columns.tolist())
print(len(dataset.columns.tolist()))
dtex.missing_percentage(data)
data.head(10)
```

```python
featureD = dtex.feature_distribution(data, 25, temporal_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold'])
```

```python
# dtex.temporal_relations(data, featureD['Temporal Features'], target_feature = 'SalePrice')
for feature in featureD['Temporal Features']:
    # plt.plot(data[feature], data['SalePrice'])
    data.groupby([feature])['SalePrice'].median().plot()
    plt.ylabel('SalePrice')
    plt.xlabel(feature)
    plt.show()
```

```python
dtex.discrete_relations(data, 'SalePrice', featureD['Discrete Features'])
```

```python
dtex.continous_plot(data, featureD['Continous Features'])
```

```python
dtex.continous_relations(data, featureD['Continous Features'], 'SalePrice')
```

```python
dtex.outliers_vis(data, featureD['Continous Features'])
```
